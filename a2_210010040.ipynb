```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import math as mat
```

# _Part A_

alpha=$a$

X component= $U cosa$

Y component= $U sina$

#### *uniform flow in y direction*

$U cosa$ = $\partial \phi $/$\partial x $ = $\partial \psi $/$\partial y $

$\phi$ = $Ux cosa$ , $\psi$ = $Uy cosa$

#### *uniform flow in x direction*

$U sina$ = - $\partial \phi $/$\partial y $ = $\partial \psi $/$\partial x $

$\phi$ =  $Uy sina$ , $\psi$ = - $Ux sina$

##### potential= $Ux cosa$ + $Uy sina$
 
##### *streamfunction=  $Uy cosa$-$Ux sina$*

$\Phi$= ($Ux cosa$ + $Uy sina$ )+ $\iota$ ($Uy cosa$-$Ux sina$)

```python

x,y=np.mgrid[-2:2:20j,-2:2:10j]
alpha=np.pi/3

u=mat.cos(alpha)
v=mat.sin(alpha)

plt.quiver(x,y,u,v)
plt.clf
plt.title('Velocity field ')
plt.show()
```

```python
x, y = np.mgrid[-2.0:2.0:50j, -2:2:50j]
phi=x*u+y*v
plt.contour(x, y, phi); 
plt.title('potential ')
```

```python
x, y = np.mgrid[-2.0:2.0:50j, -2:2:50j]
psi=x*u-y*v
plt.contour(x, y, psi);
plt.title('streamfunction')
```

# _Part B_

```python
x,y=np.mgrid[-2:2:30j,-2:2:30j]
r=np.sqrt(x**2+y**2)
u=x/(2*np.pi*r**2)
v=y/(2*np.pi*r**2)

u[(r<1)]=0
v[(r<1)]=0
plt.quiver(x,y,u,v)
plt.clf
plt.title('Velocity field of source ')
plt.show()


```

```python
x,y=np.mgrid[-2:2:30j,-2:2:30j]
r=np.sqrt(x**2+y**2)
u=-(x-3)/(2*np.pi*r**2)
v=-(y-3)/(2*np.pi*r**2)

u[(r<1)]=0
v[(r<1)]=0
plt.quiver(x,y,u,v)
plt.clf
plt.title('Velocity field of sink ')
plt.show()
```

```python
x, y = np.mgrid[-2.0:2.0:30j, -2:2:30j]
phi=x*u+y*v
plt.contour(x, y, phi);
plt.title('potential ')
```

```python
x, y = np.mgrid[-2.0:2.0:30j, -2:2:30j]
psi=x*u-y*v
plt.contour(x, y, psi);
plt.title('streamfunction')
```
